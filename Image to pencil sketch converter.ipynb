# Image to pencil sketch converter

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('doraemon.png')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
print("Normal Image")
plt.show()
```

# Conversion

```python
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(grey_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
#plt.show()
ivrt_img = 255 - grey_img
plt.imshow(cv2.cvtColor(ivrt_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
#plt.show()
blurr = cv2.GaussianBlur(ivrt_img, (21,21), 0)
plt.imshow(cv2.cvtColor(blurr, cv2.COLOR_BGR2RGB))
plt.axis('off')
#plt.show()
ivrt_blurr = 255 - blurr
pencil_sketch = cv2.divide(grey_img, ivrt_blurr, scale=256.0)
plt.imshow(cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB))
plt.axis('off')
print("Pencil Sketch")
plt.show()
```
